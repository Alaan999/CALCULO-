```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
print("Introduce la matriz de adyacencia separando elementos por espacio:")
filas = int(input("Número de nodos: "))
elementos = list(map(int, input(f"Introduce los {filas**2} elementos: ").split()))
matriz = np.array(elementos).reshape(filas, filas)
grafo = nx.from_numpy_array(matriz)

nx.draw(grafo, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```
